```python
# Import libraries
import numpy as np
import glob
from infer import Inference
```

```python
# load model
model = Inference()
```

```python
from IPython.display import Image, display
sample = 'inputs/download.jpeg'
model.predict('vit', image_path=sample, postprocess_and_save=True)

print(f"Processed image: {sample}")
display(Image(filename=sample))
```

```python
# bulk prediction
import pandas as pd
images_folder = os.path.expanduser('~/data/image_201509_onlyrefugee/')
image_paths = glob.glob(os.path.join(images_folder, "*"))
y_pred = []
error_messages = []
for path in image_paths:
    try: 
        pred = model.predict(model_name="vit", image_path=path, postprocess_and_save=False)
        y_pred.append((path, pred))
    except Exception as e:
        error_message = str(e)
        error_messages.append(error_message)
df_pred = pd.DataFrame(data=y_pred, columns=['image', 'prediction'])
```

```python
path_of_project = os.path.expanduser('~/qca-ig-posts/image_features/camera-angle/image_201509_onlyrefugee.csv')
df_pred.to_csv(path_of_project)
```
